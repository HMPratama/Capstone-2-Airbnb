### **1a. Host dan Area Paling banyak Review**

```python
host_areas = data_airbnb.groupby(['host_name','area'])[['calculated_host_listings_count','room_type','number_of_reviews_ltm','last_review']].max().reset_index()
host_areas_TOP10=host_areas.sort_values(by='number_of_reviews_ltm',ascending=False).head(10)
host_areas_TOP10

#number_of_reviews_ltm : number of review yang sudah ada selama 12 terakhir
```

```python
# Set up the figure with subplots
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Plot the first bar plot
sns.barplot(ax=axes[0], data=host_areas_TOP10, x='host_name', y='number_of_reviews_ltm')
axes[0].set_xlabel('Host Name', fontsize=12)
axes[0].set_ylabel('Number of Reviews', fontsize=12)
axes[0].set_title('Room by host with higher Reviews', fontsize=14)
axes[0].tick_params(axis='x', rotation=45)

# Plot the second bar plot
sns.barplot(ax=axes[1], data=host_areas_TOP10, x='area', y='number_of_reviews_ltm')
axes[1].set_xlabel('Area', fontsize=12)
axes[1].set_ylabel('Number of Reviews', fontsize=12)
axes[1].set_title('Area by host with higher Reviews', fontsize=14)
axes[1].tick_params(axis='x', rotation=45)

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```

* Dari visualisasi diatas terlihat bahwa host name **Paw** memiliki number of review paling tinggi setelah host name yang lainnya.
* Untuk daerah yang paling banyak di kunjungi **Ratchathewi**
* Mengapa review itu penting, menurut https://www.airbnb.co.id/resources/hosting-homes/a/why-reviews-matter-41
* Dengan adanya review yang membantu tamu untuk memilih rencana perjalanan mereka. Data Airbnb menujukan bahwa tamu lebih mungkin memesan suatu tempat tersebut meiliki review yang lebih tinggi atau lebih banyak.

### **Bagaimana ketersediaan properti berdasarkan data Airbnb**

```python
plt.rcParams['figure.figsize'] = (10, 10)

top_10_areas = data_airbnb_1['area'].value_counts().head(10).index
data_subset = data_airbnb_1[data_airbnb_1['area'].isin(top_10_areas)].sort_values(by='area',ascending=False)
ax= sns.countplot(y='room_type',data=data_subset, hue='area',palette='bright')
total = len(data_airbnb_1['room_type'])
for p in ax.patches:
        percentage = '{:.1f}%'.format(100 * p.get_width()/total)
        x = p.get_x() + p.get_width() + 0.02
        y = p.get_y() + p.get_height()/2
        ax.annotate(percentage, (x, y))

plt.title('Count of each room types each Area')
plt.xlabel('Rooms')
plt.xticks(rotation=90)
plt.ylabel('Room Counts')

plt.show()
```

jumlah populasi properti yang berdasarkan room type:
